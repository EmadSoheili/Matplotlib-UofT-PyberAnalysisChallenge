### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline

# Dependencies and Setup
import os
import pandas as pd
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import style
import matplotlib.dates as mdates


# File to Load (Remember to change these)
city_data_to_load = os.path.join("Resources","city_data.csv")
ride_data_to_load = os.path.join("Resources","ride_data.csv")

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
ride_byCityType = pyber_data_df.groupby('type').count()['ride_id']
ride_byCityType
```

```python
# 2. Get the total drivers for each city type
driver_byCityType = city_data_df.groupby('type').sum()['driver_count']
driver_byCityType
```

```python
#  3. Get the total amount of fares for each city type
fare_byCityType = pyber_data_df.groupby('type').sum()['fare']
fare_byCityType
```

```python
#  4. Get the average fare per ride for each city type. 
avgFare_perRide_byCityType = pyber_data_df.groupby('type').mean()['fare']
avgFare_perRide_byCityType
```

```python
# 5. Get the average fare per driver for each city type. 
avgFare_perDriver_byCityType = pyber_data_df.groupby('type').sum()['fare'] / city_data_df.groupby('type').sum()['driver_count']
avgFare_perDriver_byCityType
```

```python
#  6. Create a PyBer summary DataFrame.
pyberSummary_df = pd.DataFrame()
pyberSummary_df['Total Rides'] = ride_byCityType
pyberSummary_df['Total Drivers'] = driver_byCityType
pyberSummary_df['Total Fares'] = fare_byCityType
pyberSummary_df['Average Fare per Ride'] = avgFare_perRide_byCityType
pyberSummary_df['Average Fare per Driver'] = avgFare_perDriver_byCityType
pyberSummary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyberSummary_df.index.name = None
pyberSummary_df
```

```python
#  8. Format the columns.
pyberSummary_df = pyberSummary_df.style.format({'Total Rides':'{:,.0f}',
                                                'Total Drivers':'{:,.0f}',
                                                'Total Fares':'${:,.2f}',
                                                'Average Fare per Ride':'${:,.2f}',
                                                'Average Fare per Driver':'${:,.2f}'
                                               })
pyberSummary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_byType_byDate = pyber_data_df.groupby(['type','date']).sum()[['fare']]
fare_byType_byDate
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fare_byType_byDate = fare_byType_byDate.reset_index()
fare_byType_byDate
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
fare_pivot = fare_byType_byDate.pivot(index='date', columns='type', values='fare')
fare_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
fare_pivot_selectedDates = fare_pivot.loc['2019-01-01':'2019-04-29']
fare_pivot_selectedDates
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fare_pivot_selectedDates.index = pd.to_datetime(fare_pivot_selectedDates.index)
fare_pivot_selectedDates
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
fare_pivot_selectedDates.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_weekly = fare_pivot_selectedDates.resample('W').sum()
fare_weekly
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

fig, ax = plt.subplots(figsize = (20,6))
ax.plot(fare_weekly['Rural'], label = 'Rural')
ax.plot(fare_weekly['Suburban'], label = 'Suburban')
ax.plot(fare_weekly['Urban'], label = 'Urban')
ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare ($USD)")
ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=(1)))
ax.xaxis.set_minor_locator(mdates.MonthLocator([2,3,4]))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b\n%Y'))
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%b'))
ax.legend(fontsize=14, loc="center", title="City Types")

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

plt.tight_layout()
plt.savefig('Analysis/PyBer_fare_summary.png')
plt.show()
```
